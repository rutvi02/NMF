```python
import numpy as np 
file_path_u1 = 'Updated_399C1-m76_5L320.230.800.D.raw'
with open(file_path_u1, 'rb') as file:
        raw_data_u2 = np.fromfile(file, dtype=np.uint8)
```

```python
width = 320
height = 230
num_frames = 399
data_length = int(width) * int(height) * int(num_frames)  # Total length of the 1D array
        
        
time_series_2d_image_data = raw_data_u2.reshape(int(num_frames), int(height), int(width))

        #time_series_2d_image_data
        
avg_intensity = []
for frame in time_series_2d_image_data:
    avg_intensity.append(np.mean(frame))

import matplotlib.pyplot as plt

# Define the data
x = []
for i in range(1,400):
    x.append(i)
print(min(avg_intensity))
print(max(avg_intensity))

print(np.mean(avg_intensity))

# Create the plot
plt.plot(x,avg_intensity)
# plt.xlim(0, 500)  # Set x-axis limits
# plt.ylim(0,180)
# Add labels and title
plt.xlabel('time frame')
plt.ylabel('Mean')
plt.title('Plot Title')

# Display the plot
plt.show()

```

```python
width = 320
height = 230
num_frames = 399
data_length = int(width) * int(height) * int(num_frames)  # Total length of the 1D array
        
        
time_series_2d_image_data = raw_data_u2.reshape(int(num_frames), int(height), int(width))

        #time_series_2d_image_data
        
avg_intensity = []
for frame in time_series_2d_image_data:
    avg_intensity.append(np.mean(frame))
```

```python
import numpy as np

# Your time series data as a 3D array (num_frames, height, width)
time_series_data = time_series_2d_image_data  # Example data

# Define a Z-score threshold for outliers
z_threshold = 1.3

avg_intensity = []
for frame in time_series_data:
    avg_intensity.append(np.mean(frame))
# Calculate the mean and standard deviation for the entire dataset
mean = np.mean(avg_intensity)
std = np.std(avg_intensity)

# Calculate Z-scores for all frames
z_score = []
for frame in time_series_data:
    #avg_intensity.append(np.mean(frame))
    a = np.mean(frame)
    z_score.append((a - mean) / std)

# Find frames with Z-scores exceeding the threshold
outlier_frames = np.where(np.abs(z_score) > z_threshold)
outlier_frames
# Remove outlier frames
time_series_data_clean = np.delete(time_series_data, outlier_frames, axis=0)
```
