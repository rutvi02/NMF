```python
###MAIN
import glob
import os
import matplotlib.pyplot as plt
import numpy as np
directory_path = '\data'
file_pattern = '*.raw'
full_pattern = os.path.join(directory_path, file_pattern)

# Get a list of matching files
file_list = glob.glob(file_pattern)
lst = []
# file_list
for file_path in file_list:
        with open(file_path, 'rb') as file:
            raw_data = np.fromfile(file, dtype=np.uint8)
        if len(file_path)== 27:
            if file_path[9] == 'L' or file_path[9] == 'R' or file_path[9] == ' ':
                width = file_path[10:13]
                height = file_path[14:17]
                num_frames = file_path[18:21]
            else:
                width = file_path[9:12]  
                height = file_path[13:16] 
                if file_path[17] == 'Z':
                    num_frames = file_path[18:21]
                else:
                    num_frames = file_path[17:21]
            #print(width,height,num_frames)
        if len(file_path) == 30:
            if file_path[9] == '_':
                width = file_path[10:13]
                height = file_path[15:18]
                num_frames = file_path[20:23]
            else:
                width = file_path[9:12]  
                height = file_path[14:17]  
                num_frames = file_path[19:22] 
            #print(width,height,num_frames)
        if len(file_path) == 35:
            width = file_path[10:13]  
            height = file_path[15:18]  
            num_frames = file_path[20:23]
            #print(width,height,num_frames)
        if len(file_path) == 28:
            if (file_path[8] == 'R' or file_path[17] == 'Z') and (file_path[9] != ' '):
                width = file_path[9:12]  
                height = file_path[13:16] 
                num_frames = file_path[18:22]
            elif file_path[9] == ' ':
                width = file_path[10:13]  
                height = file_path[14:17] 
                num_frames = file_path[18:22] 
            else:               
                width = file_path[10:13]  
                height = file_path[14:17]
                if file_path[18] == 'Z':
                    num_frames = file_path[19:22]
                elif file_path[8] == 'L': # CHANGE
                    num_frames = file_path[18:22]
                else:
                    num_frames = file_path[18:21]
            #print(width,height,num_frames)
        if len(file_path) == 26:
            width = file_path[9:12]  
            height = file_path[13:16]  
            num_frames = file_path[17:20]
            #print(width,height,num_frames)
        if len(file_path) == 25:
            width = file_path[9:12]  
            height = file_path[13:16]  
            num_frames = file_path[18:21]
            #print(width,height,num_frames)
        
        data_length = int(width) * int(height) * int(num_frames)  # Total length of the 1D array
        
        
        time_series_2d_image_data = raw_data.reshape(int(num_frames), int(height), int(width))

        #time_series_2d_image_data
        
        avg_intensity = []
        for frame in time_series_2d_image_data:
            avg_intensity.append(np.mean(frame))
            
        max_index = avg_intensity.index(max(avg_intensity))
        
        avg_intensity_diff = [0]*len(avg_intensity)
        avg_intensity_diff.append(avg_intensity[0])
        for i in range(1,max_index+15):
            avg_intensity_diff[i] = avg_intensity[i] - avg_intensity[i-1]
        
        max_diff = max(avg_intensity_diff)
        max_diff_index = avg_intensity_diff.index(max_diff)
        
        new_array = time_series_2d_image_data[0:max_diff_index-2]
        
        
                
#         plt.imshow(new_array[-1], cmap='gray')  # Assuming grayscale images
#         plt.title(f"Frame {file_path}")
#         plt.axis('off')  # Turn off axis labels
#         plt.show()
        
#         print(max_diff_index)
#         if file_path == 'C1-m20_4R300.138.1000.N.raw':
#             print(file_path)
            
        
        #save the changed file 
        one_d_array = new_array.flatten()
        
        file_output_path = 'Updated' + '_' + str(max_diff_index-2) + file_path 

        # Save the array to the .raw file
        one_d_array.tofile(file_output_path)
        print('Loaded')
        
```
